```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

review_top_business['punct'] = review_top_business['text'].apply(lambda x: remove_punct(x))
review_top_business.head(10)
```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

review_top_business['tokenized'] = review_top_business['punct'].apply(lambda x: tokenization(x.lower()))
review_top_business.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
review_top_business['nonstop'] = review_top_business['tokenized'].apply(lambda x: remove_stopwords(x))
review_top_business.head(10)
```

```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

review_top_business['stemmed'] = review_top_business['nonstop'].apply(lambda x: stemming(x))
review_top_business.head()
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

review_top_business['lemmatized'] = review_top_business['nonstop'].apply(lambda x: lemmatizer(x))
review_top_business.head()
```

```python
def clean_text(text):
    text_lc = "".join([word.lower() for word in text if word not in string.punctuation]) # remove puntuation
    text_rc = re.sub('[0-9]+', '', text_lc)
    tokens = re.split('\W+', text_rc)    # tokenization
    text = [ps.stem(word) for word in tokens if word not in stopword]  # remove stopwords and stemming
    return text
```

```python
def get_top_n_gram(corpus,ngram_range,n=None):
    vec = CountVectorizer(ngram_range=ngram_range,stop_words = 'english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```
